````python
def ProperCase(s):
    """
    Returns a properly cased string.
    - First character is uppercase
    - All remaining characters are lowercase
    Example: "jOHN" becomes "John"
    """
    return s[0].upper() + s[1:].lower()


def RemoveNewLine(s):
    """
    Removes newline characters from a string.
    Uses the replace() method to substitute "\n" with an empty string.
    Example: "John\n" becomes "John"
    """
    return s.replace("\n", "")


def Trim(s):
    """
    Removes leading and trailing spaces from a string.
    Uses the strip() method to clean up whitespace at both ends.
    Example: "  John  " becomes "John"
    """
    return s.strip()


def FirstName(s):
    """
    Extracts and returns the first name from a full name string.
    Steps:
    1. Find the position of the first space using find()
    2. Extract substring from start to first space
    3. Apply ProperCase() for proper formatting
    Example: "john Q. Public" returns "John"
    """
    space_pos = s.find(" ")
    return ProperCase(s[:space_pos])


def LastName(s):
    """
    Extracts and returns the last name from a full name string.
    Steps:
    1. Find the position of the last space using rfind()
    2. Extract substring from last space to end (skip the space)
    3. Apply ProperCase() for proper formatting
    Example: "john Q. Public" returns "Public"
    """
    space_pos = s.rfind(" ")
    return ProperCase(s[space_pos + 1:])


def MiddleName(s):
    """
    Extracts and returns the middle name/initial from a full name string.
    Steps:
    1. Find the first space position using find()
    2. Find the last space position using rfind()
    3. Extract substring between first and last space
    4. Apply Trim() to remove extra spaces
    5. Apply ProperCase() for proper formatting
    6. If the result is a single character (initial), append a period
    Example: "john Q. Public" returns "Q."
    """
    first_space = s.find(" ")
    last_space = s.rfind(" ")
    middle = Trim(s[first_space + 1:last_space])
    middle = ProperCase(middle)
    if len(middle) == 1:
        middle += "."
    return middle


# Main Program
# Open and process the names.txt file
print(f"{'First':<10} | {'Middle':<10} | {'Last':<10}")
print("-" * 35)

try:
    with open("names.txt", "r") as file:
        for line in file:
            # Clean the input: remove newline and trim spaces
            clean_name = Trim(RemoveNewLine(line))

            # Extract name components
            first = FirstName(clean_name)
            middle = MiddleName(clean_name)
            last = LastName(clean_name)

            # Print formatted output
            print(f"{first:<10} | {middle:<10} | {last:<10}")
except FileNotFoundError:
    print("Error: names.txt file not found in the current directory.")
````

````python
def ProperCase(s):
    """
    Returns a properly cased string.
    - First character is uppercase
    - All remaining characters are lowercase
    Example: "jOHN" becomes "John"
    """
    if not s:  # Handle empty string case
        return ""
    return s[0].upper() + s[1:].lower()


def RemoveNewLine(s):
    """
    Removes newline characters from a string.
    Uses the replace() method to substitute "\n" with an empty string.
    Example: "John\n" becomes "John"
    """
    return s.replace("\n", "")


def Trim(s):
    """
    Removes leading and trailing spaces from a string.
    Uses the strip() method to clean up whitespace at both ends.
    Example: "  John  " becomes "John"
    """
    return s.strip()


def FirstName(s):
    """
    Extracts and returns the first name from a full name string.
    Steps:
    1. Find the position of the first space using find()
    2. Extract substring from start to first space
    3. Apply ProperCase() for proper formatting
    Example: "john Q. Public" returns "John"
    """
    space_pos = s.find(" ")
    # If no space is found, the whole string is considered the first name
    if space_pos == -1:
        return ProperCase(s)
    return ProperCase(s[:space_pos])


def LastName(s):
    """
    Extracts and returns the last name from a full name string.
    Steps:
    1. Find the position of the last space using rfind()
    2. Extract substring from last space to end (skip the space)
    3. Apply ProperCase() for proper formatting
    Example: "john Q. Public" returns "Public"
    """
    space_pos = s.rfind(" ")
    # If no space is found, there might not be a last name or it's a single word name
    if space_pos == -1:
        return ""
    return ProperCase(s[space_pos + 1:])


def MiddleName(s):
    """
    Extracts and returns the middle name/initial from a full name string.
    Steps:
    1. Find the first space position using find()
    2. Find the last space position using rfind()
    3. Extract substring between first and last space
    4. Apply Trim() to remove extra spaces
    5. Apply ProperCase() for proper formatting
    6. If the result is a single character (initial), append a period
    Example: "john Q. Public" returns "Q."
    """
    first_space = s.find(" ")
    last_space = s.rfind(" ")

    # Handle cases with only first and last name (no middle name)
    if first_space == last_space:
        return ""

    middle = Trim(s[first_space + 1:last_space])
    middle = ProperCase(middle)
    if len(middle) == 1:
        middle += "."
    return middle


# Main Program
# Open and process the names.txt file
print(f"{'First':<10} | {'Middle':<10} | {'Last':<10}")
print("-" * 35)

try:
    with open("names.txt", "r") as file:
        for line in file:
            # Clean the input: remove newline and trim spaces
            clean_name = Trim(RemoveNewLine(line))

            # Handle cases where the line might be empty or only contains one name
            if not clean_name:
                continue

            # Extract name components
            first = FirstName(clean_name)
            middle = MiddleName(clean_name)
            last = LastName(clean_name)

            # Print formatted output
            print(f"{first:<10} | {middle:<10} | {last:<10}")
except FileNotFoundError:
    print("Error: names.txt file not found in the current directory.")
````

# Aassignment 2
## Due March 19,  2026
* Prompt for a sequence of floating point or integer numbers, the last number being zero.
* Assume that at least one number will be entered.
* When the zero is entered, display the:
    * Count
    * Sum
    * Average
    * Minimum
    * Maximum
* Do **NOT** store the values in a list or use the min(), max(), or mean() functions

**Example Output**

```
  Enter a Value (zero to quit): 5
  Enter a Value (zero to quit): 6
  Enter a Value (zero to quit): 11
  Enter a Value (zero to quit): 0
  Count:   3
  Sum:     22.0
  Average: 7.333333333333333
  Minimum: 5.0
  Maximum: 11.0
```

# Grading Rubric
* 10 points - Count
* 10 points - Sum
* 10 points - Average
* 10 points - Minimum
* 10 points - Maximum

````python
count = 0
T_Sum = 0
min_ = None
max_ = None

def get_float(prompt):
    while True:
        try:
            user_input = input(prompt)
            if user_input == '': # Handle empty input directly
                print("Input cannot be empty. Please enter a value.")
                continue
            return float(user_input)
        except ValueError:
            print("Invalid input. Please enter a valid number.")

# Prompt for numerical values, until 0
while True:
    value = get_float("Enter a Value (zero to quit): ")

    if value == 0:
        # Ensure at least one number was entered before breaking if required by assignment
        if count == 0:
            print("At least one number must be entered before zero.")
            continue
        break

    count += 1
    T_Sum += value

    # Update minimum
    if min_ is None or value < min_:
        min_ = value
    # Update maximum
    if max_ is None or value > max_:
        max_ = value

# Handle case where no valid numbers were entered (only zero immediately)
if count == 0:
    print("No numbers were entered to calculate statistics.")
else:
    # Calculate average
    avg = T_Sum / count

    # Print result
    print(f"Count:   {count}")
    print(f"Sum:     {T_Sum}")
    print(f"Average: {avg}")
    print(f"Minimum: {min_}")
    print(f"Maximum: {max_}")
````

```python
# Unit Conversion

# Prompt for FromValue
def get_float(prompt):
    try:
        return float(input(prompt))
    except ValueError:
        print("Error: You must enter numeric value.")
        exit()
FromValue = get_float("Enter from value: ")

# Prompt for FromUnit
FromUnit = input("Enter from Unit(in, ft, yd, mi): ")

# Validate FromUnit
if FromUnit not in ['in', 'ft', 'yd', 'mi']:
    print("FromUnit is not valid")
    exit()

# Prompt for ToUnit
ToUnit = input("Enter to Unit(in, ft, yd, mi): ")

# Validate ToUnit
if ToUnit not in ['in', 'ft', 'yd', 'mi']:
    print("ToUnit is not valid")
    exit()

# Determine multiplier
# Explicitly define all 16 combinations as separate if statements

# Same unit conversions
if FromUnit == 'in' and ToUnit == 'in': multiplier = 1.0
if FromUnit == 'ft' and ToUnit == 'ft': multiplier = 1.0
if FromUnit == 'yd' and ToUnit == 'yd': multiplier = 1.0
if FromUnit == 'mi' and ToUnit == 'mi': multiplier = 1.0

# in -> others
if FromUnit == 'in' and ToUnit == 'ft': multiplier = 1/12
if FromUnit == 'in' and ToUnit == 'yd': multiplier = 1/36
if FromUnit == 'in' and ToUnit == 'mi': multiplier = 1/63360

# ft -> others
if FromUnit == 'ft' and ToUnit == "in": multiplier = 12
if FromUnit == 'ft' and ToUnit == 'yd': multiplier = 1/3
if FromUnit == 'ft' and ToUnit == 'mi': multiplier = 1/5280

# yd -> others
if FromUnit == 'yd' and ToUnit == 'in': multiplier = 36
if FromUnit == 'yd' and ToUnit == 'ft': multiplier = 3
if FromUnit == 'yd' and ToUnit == 'mi': multiplier = 1/1760

# mi -> others
if FromUnit == 'mi' and ToUnit == 'in': multiplier = 63360
if FromUnit == 'mi' and ToUnit == 'ft': multiplier = 5280
if FromUnit == 'mi' and ToUnit == 'yd': multiplier = 1760

# Compute ToValue
ToValue =  round(FromValue * multiplier, 7)

#Display Result
print(f"{FromValue} {FromUnit} = {ToValue} {ToUnit}")
```
